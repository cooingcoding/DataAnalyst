# 비가 오면 지하철을 많이 탈까

```python
rain = pd.read_csv('/content/extremum_20230102232854.csv', encoding = 'CP949')
```

```python
rain.info()
```

```python
rain
```

결측치 처리

```python
rain  = rain.fillna(0)
```

필요한 데이터컬럼 추출

```python
rain2 = rain[['일시', '강수량(mm)']]
```

```python
rain2.columns = ['day', 'rain']
```

```python
rain2
```

```python
sub = pd.read_csv('/content/서울교통공사 2021년 일별 역별 시간대별 승하차인원(1_8호선).csv', encoding = 'CP949')
```

```python
sub.info()
```

```python
sub.head()
```

```python
sub['날짜'].nunique()
```

필요한 데이터 컬럼 뽑기

```python
sub2 = sub[['날짜', '역명', '구분', '합 계']]
```

```python
sub2.info()
```

```python
sub2
```

```python
sub2.columns = ['day', 'station', 'on_off', 'sum']
```

groupby로 일별 합계 내기

```python
sub3 = sub2.groupby('day').sum()
```

데이터 결합

```python
m_data  = pd.merge(rain2, sub3, on = 'day', how = 'inner')
```

```python
m_data
```

결과값 정리

```python
rain_data = m_data[m_data['rain'] > 0]  #비가 온 날
no_rain_data = m_data[m_data['rain'] == 0]  # 비가 안온 날
```

```python
rain_data['sum'].mean()
```

```python
no_rain_data['sum'].mean()
```

비가 많이 온다고 지하철을 많이 타지는 않음
