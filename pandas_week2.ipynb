# 1. 데이터 생성하기

```python
import pandas as pd
```

## 열 별로 데이터 생성하기

```python
df = pd.DataFrame(
{"a" : [4, 5, 6],
"b" : [7, 8, 9],
"c" : [10, 11, 12]},
index = [1, 2, 3])
```

```python
df
```

데이터 프레임의 기본 구조

1. 행과 열로 이루어짐(row & column)
2. 행과 열은 각각 인덱스, 컬럼 이름으로 접근함

## 행 별로 데이터 생성하기

```python
df = pd.DataFrame(
[[4, 7, 10],
[5, 8, 11],
[6, 9, 12]],
index=[1, 2, 3],
columns=['a', 'b', 'c'])
```

```python
df
```

## multi-index 데이터 프레임 만들기

```python
df = pd.DataFrame(
{"a" : [4 ,5, 6],
"b" : [7, 8, 9],
"c" : [10, 11, 12]},
index = pd.MultiIndex.from_tuples(
[('d', 1), ('d', 2),
('e', 2)], names=['n', 'v']))
```

```python
df
```
